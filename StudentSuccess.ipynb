## Count of students by marital status

```python
# Distribution of students by marital status
statuses = ["Single", "Married", "Widower", "Divorced", "Facto Union", "Legally Separated"]
# One Hot Encoding Of Marital Status Feature
marital_status = data["Marital status"]
marital_status = pd.get_dummies(marital_status,dtype=int)
marital_status.columns = statuses
marital_status.head()
# data.groupby("Marital status")["Marital status"].count()
# sns.barplot(data = data.groupby("Marital status")["Marital status"].count() )
```
